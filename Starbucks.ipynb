```python
# Importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading file of interest

Starbucks = pd.read_csv("starbucks_drinkMenu_expanded.csv")
Starbucks.head()
```

```python
# Quick glance through the dataset revealed that there are 18 columns with about 242 entries

Starbucks.info()
```

```python
Starbucks.describe()
```

A statistic summary of the columns of this dataset shows:

1. Most of the beverages on Starbucks menu has more calories with high total caborhydrates being a major source of most of the calories found in the beverages.
2. The nutrients that were found the least in all the beverages are Saturated Fat followed  by dietary fibre and Trans fat.
   

```python
# Viewing the distributions of the features

Starbucks.hist(figsize=(15,10))
plt.subplots_adjust(hspace =0.5)
```

The histograms reveal that the four most abundant nutrients in Starbucks bevearges are Total Carbohydrates, Cholesterol, Sugars and Protein

```python
# Checking the average number of calories per beverage

fig,ax= plt.subplots(figsize=(23,16))
sns.boxplot(y= 'Beverage', x= 'Calories', data = Starbucks)
```

This boxplot shows that the beverage with the most calories is java chi[(without whipped cream) and the beverage with the least calories is Tazo Tea.
Caffè Latte has an average of  about 150 calories 12oz serving.

```python
# Checking how these beverages are prepared and  how the prep contributes to the number of calories

sns.barplot(y= 'Beverage_prep', x= 'Calories', data = Starbucks)
```

Beverages prepared with whole milk tend to have more Calories.

```python
# Checking the different ways beverages can be prepared

fig,ax= plt.subplots(figsize=(24,16))
sns.scatterplot(x= 'Beverage_prep', y= 'Beverage', data = Starbucks)
```

From the scatterplot, Caffè Latte can be prepared in 6 ways with: 
1. Short Nonfat Milk
2. 2% Milk
3. Soy milk
4. Tall Nonfat milk
5. Grande Nonfat Milk
6. Venti Nonfat milk

These different ways of prep  determines how many calories are present in the beverage.

```python
# Viewing nutritional information of Caffè Latte prepares in different ways

High_cal= Starbucks[Starbucks['Beverage']=='Caffè Latte']
High_cal
```

```python
fig,ax= plt.subplots(figsize=(12,8))
sns.barplot(y= 'Beverage_prep', x= 'Calories', data = High_cal)
plt.title("Different preparations of Caffè Latte and its equivalent amount of Calories")	 
```

Preparing Caffè Latte with 2% Milk increases the Calories content the most(240). While preparing Caffè Latte with Short Nonfat milk reduces the calories content to about 70.
